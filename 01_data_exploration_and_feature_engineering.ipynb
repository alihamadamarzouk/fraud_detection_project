```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python

labels = pd.read_csv("/content/Train-1542865627584.csv")
bene   = pd.read_csv("/content/Train_Beneficiarydata-1542865627584.csv")
inp    = pd.read_csv("/content/Train_Inpatientdata-1542865627584.csv")
outp   = pd.read_csv("/content/Train_Outpatientdata-1542865627584.csv")
```

```python
print("Beneficiary:", bene.shape)
display(bene.head())

print("Inpatient:", inp.shape)
display(inp.head())

print("Outpatient:", outp.shape)
display(outp.head())

print("Labels:", labels.shape)
display(labels.head())
```

```python
print("Inpatient missing values:")
print(inp.isna().sum().sort_values(ascending=False).head())

print("\nOutpatient missing values:")
print(outp.isna().sum().sort_values(ascending=False).head())
```

```python
labels["target"] = labels["PotentialFraud"].map({"Yes":1, "No":0})
labels["target"].value_counts()
```

```python
labels["target"].value_counts().plot(kind="bar")
plt.title("Fraud vs Not Fraud (Providers)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
inpatient_features = inp.groupby("Provider").agg({
    "ClaimID": "count",
    "InscClaimAmtReimbursed": "sum"
}).rename(columns={
    "ClaimID": "Inpatient_Claims",
    "InscClaimAmtReimbursed": "Inpatient_TotalAmount"
})
```

```python
outpatient_features = outp.groupby("Provider").agg({
    "ClaimID": "count",
    "InscClaimAmtReimbursed": "sum"
}).rename(columns={
    "ClaimID": "Outpatient_Claims",
    "InscClaimAmtReimbursed": "Outpatient_TotalAmount"
})
```

```python
bene['DOB'] = pd.to_datetime(bene['DOB'])

# Calculate Age based on a fixed reference date (end of 2009)
REFERENCE_DATE_FOR_AGE = pd.to_datetime('2009-12-31')
bene['Age'] = (REFERENCE_DATE_FOR_AGE - bene['DOB']).dt.days / 365.25

# Identify all chronic condition columns
chronic_cols = [col for col in bene.columns if 'ChronicCond_' in col]

# Create a binary column 'HasChronicCondition': 1 if any chronic condition is present (value 1), 0 otherwise
# Chronic condition columns are coded as 1 for 'Yes' and 2 for 'No'.
bene['HasChronicCondition'] = (bene[chronic_cols] == 1).any(axis=1).astype(int)

bene_inp = inp[["BeneID", "Provider"]].merge(bene, on="BeneID", how="left")
bene_outp = outp[["BeneID", "Provider"]].merge(bene, on="BeneID", how="left")
bene_all = pd.concat([bene_inp, bene_outp])

beneficiary_features = bene_all.groupby("Provider").agg(
    Avg_Patient_Age=("Age", "mean"),
    Unique_Patients=("BeneID", pd.Series.nunique),
    # Use the new 'HasChronicCondition' column for the percentage calculation
    Chronic_Patient_Perc=("HasChronicCondition", lambda x: np.mean(x) * 100)
).reset_index()
```

```python
provider_df = inpatient_features.merge(outpatient_features, on="Provider", how="outer").fillna(0)

provider_df = provider_df.merge(
beneficiary_features, on="Provider", how="left"
).fillna(0)

provider_df = provider_df.merge(
labels[["Provider","target"]],
on="Provider",
how="left"
).fillna(0)

provider_df.head()
```

```python
provider_df.to_csv("provider_features.csv", index=False)
print("Saved provider_features.csv with beneficiary features")
```
